```python
class Parent:
    def __init__(self, name):
        self.name = name


class Child:
    def __init__(self, name, age):
        super.__init__(name)
        self.age = age
        
```

```python
class A:
    def process(self):
        print("A's process")

class B(A):
    def process(self):
        print("B's process")
        super().process()

class C(A):
    def process(self):
        print("C's process")
        super().process()

class D(B, C):
    def process(self):
        print("D's process")
        super().process()

d = D()
d.process()
```
