## HASHING

```python
arr = [5, 6, 5, 6, 9, 6]
count = 0
for num in arr:
    if num == 6:
        count += 1
print(count)  # Output: 3
```

## Method 2: Hashing for Efficient Counting

```python
arr = [5, 6, 5, 6, 9, 6]
hash_table = [0] * 10
print(hash_table)
for num in arr:
    hash_table[num] += 1
print(hash_table[6])  # Output: 3
print(hash_table)
```
