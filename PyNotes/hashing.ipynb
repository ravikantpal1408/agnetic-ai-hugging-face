## HASHING

```python
arr = [5, 6, 5, 6, 9, 6]
count = 0
for num in arr:
    if num == 6:
        count += 1
print(count)  # Output: 3
```

## Method 2: Hashing for Efficient Counting

```python
arr = [5, 6, 5, 6, 9, 6]
hash_table = [0] * 10
print(hash_table)
for num in arr:
    hash_table[num] += 1
print(hash_table[6])  # Output: 3
print(hash_table)
```

```python

def mostFrequentElement(nums):
    
    # Variable to store the size of array
    n = len(nums)
    
    # Variable to store maximum frequency
    maxFreq = 0 
    
    # Variable to store element 
    # with maximum frequency
    maxEle = 0
    
    # Visited array
    visited = [False] * n
    
    # First loop
    for i in range(n):
        # Skip second loop if already visited
        if visited[i]:
            continue
        
        # Variable to store frequency
        # of current element 
        freq = 0
        
        # Second loop
        for j in range(i, n):
            if nums[i] == nums[j]:
                freq += 1
                visited[j] = True
        
        # Update variables if new element having 
        # highest frequency is found
        if freq > maxFreq:
            maxFreq = freq
            maxEle = nums[i]
        elif freq == maxFreq:
            maxEle = min(maxEle, nums[i])
    
    # Return the result
    return maxEle

nums = [1, 2, 2, 3, 3, 3]

result = mostFrequentElement(nums)

print(f"Most frequent element := {result}")
```
