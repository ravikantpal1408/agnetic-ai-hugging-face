```python
#Create a calculator (+, -, *, /).

def calculator(a, b , opt):
    if opt == '+':
        print(f'Sum: {a+b}')
    elif opt == '-':
        print(f'Subtraction: {a-b}')
    elif opt == '*':
        print(f'Product: {a*b}')
    elif opt == '/':
        print(f'Divide: {a/b}')
    else:
        pass


a = input('Enter first number\n')
b = input('Enter second number\n')
opt = input('Enter operator [+, -, *, /]\n')

calculator(a,b,opt)

```

```python
num = input('Enter a number\n')
try:
    if int(num) % 2==0:
        print('Entered number is even')
    else:
        print('Entered number is odd')
except ValueError:
    print('Entered value is not correct value')
```
